```python
#using matplotlib to visualize the number of autistic and non-autistic patients
plt.hist(df['Class/ASD'], bins=30, edgecolor='black', alpha=0.7)
```

```python
#using matplotlib to visualize the average age of autistic and non-autistic patients
# Group by 'Class/ASD' and calculate the mean age
mean_age = df.groupby('Class/ASD')['age'].mean().reset_index()


bars = plt.bar(mean_age['Class/ASD'], mean_age['age'], color=['blue', 'green'])  # Colors for Autistic and Non-Autistic

# Add labels and title
plt.title('Average Age of Autistic and Non-Autistic Patients')
plt.xlabel('Class/ASD')
plt.ylabel('Average Age')
plt.ylim(0, mean_age['age'].max() + 5)  # Set y-axis limit slightly above the maximum mean age

# Annotate each bar with its value
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Show plot
plt.show()
```

```python
# Count the occurrences of each gender
gender_count = df['gender'].value_counts()

bars = plt.bar(gender_count.index, gender_count.values, color=['blue', 'pink'])  # Colors for Male and Female

# Add labels and title
plt.title('Count of Males and Females')
plt.xlabel('Gender')
plt.ylabel('Count')

# Annotate each bar with its value
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Show plot
plt.show()
```

```python
# Group by 'Class/ASD' and 'gender' and calculate the count
gender_count = df.groupby(['Class/ASD', 'gender']).size().reset_index(name='count')

# Get unique genders
genders = gender_count['gender'].unique()

# Plot each gender's count for each class
for i, gender in enumerate(genders):
    subset = gender_count[gender_count['gender'] == gender]
    plt.bar(subset['Class/ASD'], subset['count'], label=gender, color=['blue', 'pink'][i])

# Add labels and title
plt.title('Count of Males and Females of Autistic and Non-Autistic Patients')
plt.xlabel('Class/ASD')
plt.ylabel('Count')
plt.legend(title='Gender')

# Annotate each bar with its value
for i, gender in enumerate(genders):
    subset = gender_count[gender_count['gender'] == gender]
    for index, value in enumerate(subset['count']):
        plt.text(index + i * 0.2 - 0.1, value + 0.2, str(value), ha='center', va='bottom')

# Show plot
plt.show()
```

```python
# Filter dataframe to include only autistic patients
filtered_df = filtered_df = df[df['Class/ASD'] == "YES"]

# Count the occurrences of each country
country_count = filtered_df['contry_of_res'].value_counts()

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(country_count.index, country_count.values, color='blue')

# Add labels and title
plt.title('Count of Autistic Patients in Different Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Annotate each bar with its value
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Group by 'Class/ASD' and 'age_desc' and calculate the count
age_count = df.groupby(['Class/ASD', 'age_desc']).size().reset_index(name='count')

# Get unique age descriptions
age_descs = age_count['age_desc'].unique()

# Plot each age description's count for each class
for i, age_desc in enumerate(age_descs):
    subset = age_count[age_count['age_desc'] == age_desc]
    plt.bar(subset['Class/ASD'], subset['count'], label=age_desc, color=['blue', 'green'][i])

# Add labels and title
plt.title('Count of People over 18 for Autistic and Non-Autistic Patients')
plt.xlabel('Class/ASD')
plt.ylabel('Count')
plt.legend(title='Age Description')

# Annotate each bar with its value
for i, age_desc in enumerate(age_descs):
    subset = age_count[age_count['age_desc'] == age_desc]
    for index, value in enumerate(subset['count']):
        plt.text(index + i * 0.2 - 0.1, value + 0.2, str(value), ha='center', va='bottom')

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Count the occurrences of each category
jundice_count = df['jundice'].value_counts()

# Create pie chart
plt.pie(jundice_count, labels=jundice_count.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightskyblue'])
plt.title('Count of People with Jaundice')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show plot
plt.show()
```

```python
# Group by 'Class/ASD' and 'jundice' and calculate the count
jundice_count = df.groupby(['Class/ASD', 'jundice']).size().reset_index(name='count')

# Get unique jundice values
jundice_values = jundice_count['jundice'].unique()

# Plot each jundice value's count for each class
for i, j in enumerate(jundice_values):
    subset = jundice_count[jundice_count['jundice'] == j]
    plt.bar(subset['Class/ASD'], subset['count'], label=j, color=['blue', 'green'][i])

# Add labels and title
plt.title('Count of people with jundice for Autistic and Non-Autistic Patients')
plt.xlabel('Class/ASD')
plt.ylabel('Count')
plt.legend(title='Jundice')

# Annotate each bar with its value
for i, j in enumerate(jundice_values):
    subset = jundice_count[jundice_count['jundice'] == j]
    for index, value in enumerate(subset['count']):
        plt.text(index + i * 0.2 - 0.1, value + 0.2, str(value), ha='center', va='bottom')

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
